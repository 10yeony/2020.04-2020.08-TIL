```python
#iloc 인덱스로...
print(df1.iloc[0:2,0:3]) #,를 기준으로 앞에는 행이 오고 뒤에는 열이 옴
print('*'*30)

print(df1.iloc[:2,:3])
print('*'*30)

#loc 라벨로...
print(df1.loc[:1, 'name':'age']) #라벨로 인식하므로 뒤에꺼 포함시킴.
print('*'*30)

print(df1.iloc[0:2]) #열을 안 써주니까 전부 다 나오는 것.
print(df1.loc[0:1]) #위와 같은 것
```

```python
# 스칼라값 가져오기
# loc, iloc는 스칼라값을 가져오는 용도는 아니다.

print(df1.iloc[1:3])
print(df1.loc[1:2])
print('*'*30)

#iat와 at로 스칼라값 가져오기
print(df1.iat[1,1]) #인덱스
print(df1.at[1,'addr']) #라벨
print(df1.iat[3,2])
print(df1.at[3,'age'])
df1.at[3,'age']=68 #수정
df1
```

### 알아두면 유용한 함수들...

```python
# 1. unique()....중복 제거... SQL에서의 DISTINCT와 동일한 기능
print(df3['day']) #중복된 데이터 함께 보임
print('*'*30)
print(df3['day'].unique())
```

```python
# 2. 각각의 컬럼들에 해당하는 값의 개수
print(df3['day'].count()) #전체 개수
df3['day'].value_counts()
```

```python
# 3. isin() ---- 
'''
unique() 함수를 이용해서 우선 어떤 요일이 있는지부터 확인
그 중에서
토, 일요일에 있는 데이터들만 검색....
isin(['Sat','Sun'])
'''
print(df3['day'].unique())
print('*'*30)
print(df3['day'].isin(['Thur','Fri']))
print('*'*30)
print(df3[df3['day'].isin(['Thur','Fri'])])
df3[df3['day'].isin(['Thur'])]
```

```python
# 4. isnull().... 누락데이터 확인하기... NaN...
print(df3.isnull().sum())
df3.isnull().sum(axis=1) #맨 마지막만 누락데이터 있는 것을 확인
```
