```python
import torch
import torchvision
import torch.nn as nn
import torchvision.transforms as transforms
import matplotlib as plt
```

### CIFAR10 DataSet Loading and Show Image
    이 코드에서는 CIFAR10 데이터셋을 로딩하는 2가지 방법에 대해서 알아본다.

```python
train_dataset = torchvision.datasets.CIFAR10(root='../../data/', #경로 없으면 생김
                            train=True, #Train DataSet을 다운받겠다.
                            transform=transforms.ToTensor(), #굉장히 중요 
                            # tensor 값으로 만들어줌.
                            # 0~255까지의 모든 픽셀값을 255로 나누어 0~1 사이의 값으로 스케일링함.
                            # augmentation 과정이 적극적으로 들어감. 변형해야 학습 효과가 높아지는데 그 과정은 일단은 나중에
                            download=True)

# 다운받은 데이터 이미지를 실제로 출력해보자
# 원래는 0~255 값이 나와야 하는데 ToTensor() 함수를 쓰면 스케일링이 진행됨. 
# 데이터를 안정적으로 만들어줌. Overfitting을 줄이기 위함.
print(train_dataset[0]) # R,G,B 픽셀값과 라벨값이 출력됨. #총 5만개

image, label = train_dataset[0] # 이미지값과 라벨값 따로 할당
print('image : ', image)
print('label : ', label)
print('*'*30)
print('Image Size : ', image.size()) # 채널이 맨 앞에 가 있음. 나중에 채널을 뒤로 보내는 작업 필요.
print('Image Shape : ', image.shape) # 위와 같음
```
