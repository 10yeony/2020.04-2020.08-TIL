### Plot 직선 그리기

```python
# 1. plot() : 직선, 혹은 꺾은선 형태의 그래프를 그릴 수 있다... 선형(직선) | 비선형(곡선)
# 간단한 데이터를 하나 만들어서...(x,y) 직선 혹은 꺾은선을 그려보자
plt.plot([1,2,3,4], [1,4,9,16]) # 첫번째 리스트값이 X, 두번재 리스트값이 Y
plt.show()

# 2. 그렇다면 직선은?
plt.plot([1,2,3,4], [1,2,3,4]) # 기울기를 없앰 
plt.show()

# 3. 리스트에 입력한 값은 Y축의 값으로 입력되며, X축 값은 자동으로 0에서 1씩 증가
plt.plot([10,20,30,40])
plt.show()
```

### 겹쳐진 그래프 그리기

```python
import numpy as np
```

```python
# 0부터 5 미만까지 데이터를 0.2 간격으로 출력
t = np.arange(0,5, 0.2)
t
```

```python
plt.plot(t,t,'--')
plt.show()
```

```python
plt.plot(t,t,'g--')
plt.show()
```

```python
plt.plot(t,t,'r--', t, t**2, 's') # t**2는 t^2 / s = square
plt.show()
```

```python
plt.plot(t,t,'r--', t, t**2, 'gs') # t**2는 t^2 / s = square
plt.show()
```

```python
plt.plot(t,t,'r--', t,t**2,'gs', t,t**3,'b^')
plt.show()
```

```python
# 이렇게 따로 써서 겹칠 수도 있음
plt.plot(t,t,'r--', label='Tips')
plt.plot(t,t**2,'gs', label='Population')
plt.plot(t,t**3,'b^', label='Product Layer')
plt.legend() # 이 부분을 해야 label 표식이 나타난다
plt.title('Wrapped Graphs')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid()
plt.show()
```

### 파도 모양의 그래프

```python
t = np.arange(0, 12, 0.01)
'''
기본으로 존재하는 figure가 아닌 새로운 figure를 생성하기 위해서는
plt.figure() 함수를 사용한다.
'''
plt.figure(figsize=(8,5)) #가로 8, 세로 5

plt.plot(t, np.sin(t), 'r', lw=10, label='sin') #line width (디폴트가 lw=3)
plt.plot(t, np.cos(t), 'g', label='cos')
plt.legend()

plt.xlabel('Time') # xlabel은 x축에 나타나는 것, 위의 label은 해당 그래프 자체의 label
plt.ylabel('Amplitute')
plt.title('SinCos Graph2')
plt.show()
```

### Pandas 데이터를 이용한 Visualization

```python
np.random.seed(100)
# 1. 딕셔너리 구조로 데이터 프레임 만들기
# np.arange(0,50,1) => 1이 스텝임
# np.random.randint(0,50,2) => 2는 개수임
data={'a':np.arange(50),
      'c':np.random.randint(0,50,50),
      'd':np.random.randn(50)
     }
data
```

```python
data['b']=data['a']+10*np.random.randn(50)
data['d']=np.abs(data['d'])*100
data
```

### 산점도(Scatter) 그래프 그리기

```python
'''
산점도 그래프는 scatter() 함수를 이용해서 그린다.
산점도는 x,y축에 해당하는 데이터들의 상관관계를 표시할 때 데이터들이
어떻게 흩어져 있는가를 알기 위해서 사용된다.

2개의 축을 기준으로 데이터가 얼마나 퍼져있는지(분포도)를 알 수 있기 때문에
산포도 혹은 산점도로 불린다.
'''
```

```python
# 실제로는 데이터가 방대할 때 strong과 color를 이렇게 일일이 지정할 수 없다.
plt.scatter([1,2,3,4],[10,20,30,40], s=[100,200,300,400], 
           c = ['yellow','green','magenta','red'])
plt.show()
```

```python
plt.scatter([1,2,3,4],[10,20,30,40], s=[100,200,300,500], 
            c=range(4))
plt.colorbar()
plt.show()
```

```python
plt.scatter([1,2,3,4],[10,20,30,40], s=[100,200,300,500], 
            c=range(4), cmap='jet')
plt.colorbar()
plt.show()
```

```python
# plt.scatter?
# scatter는 라벨 지정이 안 됨
# 하나의 데이터가 x인 a와 y인 b로 구성되어서 구분 못함
plt.scatter('a','b', data = data, c='c')
plt.show()
```

```python
import seaborn as sns

cuthalf = (data['a']>25) & (data['b']>30)
data['color'] = np.where(cuthalf==True, 'red', 'blue')

sns.regplot(x = data['a'],
           y = data['b'],
           scatter_kws={'facecolors':data['color']})
plt.title('Scatter Plot with Colors', fontsize=20)
plt.show()
```

```python
# fit_Reg=False 로 주면 선형회귀선을 없앨 수 있다
cuthalf = (data['a']>25) & (data['b']>30)
data['color'] = np.where(cuthalf==True, 'red', 'blue')

sns.regplot(x = data['a'],
           y = data['b'],
            fit_reg=False,
           scatter_kws={'facecolors':data['color']})
plt.title('Scatter Plot with Colors', fontsize=20)
plt.show()
```

### Subplot 으로 plot(), scatter(), bar() 동시 이용하기

```python
names = ['group_A','group_B','group_C'] #x축으로 사용
values = [1, 10, 100] #y축으로 사용
```

```python
# 그림을 그리는 새로운 영역을 지정하는 함수.. figure
plt.figure(1, figsize=(9,3)) # figsize=(9,3) 영역이 첫번째라는 의미로 1 부여

plt.subplot(131)
plt.bar(names,values)

plt.subplot(132)
plt.scatter(names,values)

plt.subplot(133)
plt.plot(names, values)

plt.show()
```
